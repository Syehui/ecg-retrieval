```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

matplotlib.rcParams['xtick.labelsize'] = "large"
matplotlib.rcParams['ytick.labelsize'] = "large"
matplotlib.rcParams['font.size'] = 20
#no idea why need subplot
fig, ax = plt.subplots(figsize=(15,12))

#set background color
plt.rcParams['axes.facecolor'] = "#ffffff"

# line 1 points
x1 = [5,10,20,30,50]
#y1 = [81.69, 79.80, 78.96, 78.79, 78.62]
y1 = [93.58, 92.90, 92.06, 91.28, 90.44]
# plotting the line 1 points 
plt.plot(x1, y1, linestyle='-', linewidth=4, marker='x', color='#308cdd', markersize=15)

#set axis label
plt.xlabel('Number of Candidates(K)', fontsize=35)
plt.ylabel('q-Micro F1', fontsize=35)

#set limit to y axis
plt.ylim((89,95))

#show annotation in everypoint
for count,(i,j) in enumerate(zip(x1,y1)):
    if count == 0: plt.annotate(str(j),xy=(i+1,j))
    elif count == 1: plt.annotate(str(j),xy=(i+0.5,j+0.1))
    elif count == 2: plt.annotate(str(j),xy=(i-1,j+0.15)) 
    elif count == 3: plt.annotate(str(j),xy=(i-1,j+0.15)) 
    elif count == 4: plt.annotate(str(j),xy=(i-1,j+0.15))


# Hide the right and top spines
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(True)
ax.spines['left'].set_visible(True)

#save to local
plt.savefig('f1.png', transparent=True)
```
